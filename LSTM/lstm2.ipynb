```python
# ============================================
# TASK 1
# TEXT PREPROCESSING
# TOKENIZATION
# SEQUENCE GENERATION
# PADDING
# ============================================

from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
import numpy as np

# ============================================
# DATASET
# ============================================

sentences = [
    "I love deep learning",
    "I love artificial intelligence",
    "Deep learning is powerful",
    "Artificial intelligence is transforming the world",
    "I enjoy learning neural networks",
    "Neural networks are inspired by the human brain"
]

# ============================================
# TOKENIZATION
# ============================================

tokenizer = Tokenizer()

tokenizer.fit_on_texts(sentences)

word_index = tokenizer.word_index

print("WORD INDEX:\n")

print(word_index)

# ============================================
# SEQUENCE GENERATION
# ============================================

sequences = tokenizer.texts_to_sequences(
    sentences
)

print("\nSEQUENCES:\n")

for seq in sequences:
    print(seq)

# ============================================
# PADDING
# ============================================

padded_sequences = pad_sequences(
    sequences,
    padding='pre'
)

print("\nPADDED SEQUENCES:\n")

print(padded_sequences)
```

```python
# ============================================
# TASK 2
# CREATE INPUT-OUTPUT SEQUENCES
# FOR NEXT WORD PREDICTION
# ============================================

input_sequences = []

# Generate sequences

for sentence in sentences:

    token_list = tokenizer.texts_to_sequences(
        [sentence]
    )[0]

    # Create n-gram sequences

    for i in range(1, len(token_list)):

        n_gram_sequence = token_list[:i+1]

        input_sequences.append(
            n_gram_sequence
        )

# ============================================
# DISPLAY SEQUENCES
# ============================================

print("INPUT SEQUENCES:\n")

for seq in input_sequences:
    print(seq)

# ============================================
# PAD INPUT SEQUENCES
# ============================================

max_sequence_len = max(
    [len(seq) for seq in input_sequences]
)

input_sequences = np.array(
    pad_sequences(
        input_sequences,
        maxlen=max_sequence_len,
        padding='pre'
    )
)

# ============================================
# CREATE X AND y
# ============================================

X = input_sequences[:, :-1]

y = input_sequences[:, -1]

print("\nX SHAPE:", X.shape)

print("y SHAPE:", y.shape)

# ============================================
# VOCAB SIZE
# ============================================

total_words = len(word_index) + 1

# ============================================
# ONE HOT ENCODING
# ============================================

from tensorflow.keras.utils import to_categorical

y = to_categorical(
    y,
    num_classes=total_words
)

print("\nONE HOT y SHAPE:", y.shape)
```

```python
# ============================================
# TASK 3
# SIMPLE RNN MODEL
# ============================================

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding
from tensorflow.keras.layers import SimpleRNN
from tensorflow.keras.layers import Dense

# ============================================
# BUILD MODEL
# ============================================

rnn_model = Sequential()

# Embedding Layer

rnn_model.add(
    Embedding(
        input_dim=total_words,
        output_dim=10,
        input_length=max_sequence_len-1
    )
)

# SimpleRNN Layer

rnn_model.add(
    SimpleRNN(
        units=100
    )
)

# Output Layer

rnn_model.add(
    Dense(
        total_words,
        activation='softmax'
    )
)

# ============================================
# COMPILE MODEL
# ============================================

rnn_model.compile(
    loss='categorical_crossentropy',
    optimizer='adam',
    metrics=['accuracy']
)

# ============================================
# MODEL SUMMARY
# ============================================

rnn_model.summary()
```

```python
# ============================================
# TASK 4
# TRAIN SIMPLE RNN MODEL
# ============================================

rnn_model.fit(
    X,
    y,
    epochs=300,
    verbose=1
)
```

```python
# ============================================
# TEST RNN PREDICTION
# ============================================

input_text = "i love"

# Convert text to sequence

token_list = tokenizer.texts_to_sequences(
    [input_text]
)[0]

# Padding

token_list = pad_sequences(
    [token_list],
    maxlen=max_sequence_len-1,
    padding='pre'
)

# Predict next word

predicted = np.argmax(
    rnn_model.predict(token_list),
    axis=-1
)

# Convert number to word

for word, index in word_index.items():

    if index == predicted:
        predicted_word = word

print("\nNEXT WORD PREDICTION:")
print(input_text, "->", predicted_word)
```

```python
# ============================================
# TASK 5
# LSTM MODEL
# ============================================

from tensorflow.keras.layers import LSTM

# ============================================
# BUILD LSTM MODEL
# ============================================

lstm_model = Sequential()

# Embedding Layer

lstm_model.add(
    Embedding(
        input_dim=total_words,
        output_dim=10,
        input_length=max_sequence_len-1
    )
)

# LSTM Layer

lstm_model.add(
    LSTM(
        units=100
    )
)

# Output Layer

lstm_model.add(
    Dense(
        total_words,
        activation='softmax'
    )
)

# ============================================
# COMPILE MODEL
# ============================================

lstm_model.compile(
    loss='categorical_crossentropy',
    optimizer='adam',
    metrics=['accuracy']
)

# ============================================
# MODEL SUMMARY
# ============================================

lstm_model.summary()
```

```python
# ============================================
# TASK 6
# TRAIN LSTM MODEL
# ============================================

lstm_model.fit(
    X,
    y,
    epochs=300,
    verbose=1
)
```

```python
# ============================================
# TEST LSTM PREDICTION
# ============================================

input_text = "i love"

# Convert to sequence

token_list = tokenizer.texts_to_sequences(
    [input_text]
)[0]

# Padding

token_list = pad_sequences(
    [token_list],
    maxlen=max_sequence_len-1,
    padding='pre'
)

# Predict next word

predicted = np.argmax(
    lstm_model.predict(token_list),
    axis=-1
)

# Convert index to word

for word, index in word_index.items():

    if index == predicted:
        predicted_word = word

print("\nNEXT WORD PREDICTION:")
print(input_text, "->", predicted_word)
```

```python
# ============================================
# TASK 7
# COMPARE RNN AND LSTM
# ============================================

# ============================================
# RNN EVALUATION
# ============================================

rnn_loss, rnn_accuracy = rnn_model.evaluate(
    X,
    y,
    verbose=0
)

print("\n================================")
print("SIMPLE RNN PERFORMANCE")
print("================================")

print("RNN Accuracy:", rnn_accuracy)
print("RNN Loss:", rnn_loss)

# ============================================
# LSTM EVALUATION
# ============================================

lstm_loss, lstm_accuracy = lstm_model.evaluate(
    X,
    y,
    verbose=0
)

print("\n================================")
print("LSTM PERFORMANCE")
print("================================")

print("LSTM Accuracy:", lstm_accuracy)
print("LSTM Loss:", lstm_loss)

# ============================================
# NEXT WORD PREDICTION COMPARISON
# ============================================

input_text = "i love"

print("\n================================")
print("NEXT WORD PREDICTION COMPARISON")
print("================================")

# ============================================
# PREPARE INPUT
# ============================================

token_list = tokenizer.texts_to_sequences(
    [input_text]
)[0]

token_list = pad_sequences(
    [token_list],
    maxlen=max_sequence_len-1,
    padding='pre'
)

# ============================================
# RNN PREDICTION
# ============================================

rnn_predicted = np.argmax(
    rnn_model.predict(token_list),
    axis=-1
)

rnn_word = ""

for word, index in word_index.items():

    if index == rnn_predicted:

        rnn_word = word

# ============================================
# LSTM PREDICTION
# ============================================

lstm_predicted = np.argmax(
    lstm_model.predict(token_list),
    axis=-1
)

lstm_word = ""

for word, index in word_index.items():

    if index == lstm_predicted:

        lstm_word = word

# ============================================
# DISPLAY RESULTS
# ============================================

print("\nInput Text:", input_text)

print("\nSimpleRNN Prediction:")
print(input_text, "->", rnn_word)

print("\nLSTM Prediction:")
print(input_text, "->", lstm_word)

# ============================================
# FINAL COMPARISON
# ============================================

print("\n================================")
print("MODEL COMPARISON")
print("================================")

if lstm_accuracy > rnn_accuracy:

    print("""
LSTM performed better because:

- Better memory handling
- Smart gates
- Better context understanding
- Improved sequence learning
""")

else:

    print("""
SimpleRNN performed similarly
because dataset is very small.
""")
```
