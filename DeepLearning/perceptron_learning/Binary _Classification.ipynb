### Perceptron Binary Classification

```python
import numpy as np
```

```python
x=np.array([1,2,3,4]) # Study Hours
y=np.array([0,0,1,1]) # Pass(1)/Fail(0)
x,y
```

### Initialise Parameters

```python
w=0.0
bias=0.0
learning_rate=1
epochs=5

print("Initial Weight and Bias are: ",w,bias)
```

### Define Activation Function

```python
def act_func(z):
    # if z>=0:
    #     return 1
    # return 0
    return 1 if z>=0 else 0
```

### Training loop

```python
for epoch in range(1,epochs+1):
    print("Epoch ",epoch)
    # Step 1 Calc Weighted Sum
    for i in range(len(x)):
        weighted_sum=w*x[i]+bias
        # Step 2 apply Activation Function
        y_pred=act_func(weighted_sum)
        # Step 3 Calc Error
        error=y[i]-y_pred
        # Step 4 Update Weight 
        w=w+learning_rate*error*x[i] # if error!=0:
            # Step 5 Update Bias
        bias=bias+learning_rate*error

        print(f"Input: {x[i]} Predicted value: {y_pred} weight: {w} Bias: {bias}")
```

```python
print("Final Weight and Bias",w,bias)
```

### Testing the model

```python
test_value=3.1
z=w*test_value+bias
y_pred=act_func(z)
print(f"For Studyhours {test_value} Predicted value: {y_pred}")
test_value=2.9
z=w*test_value+bias
y_pred=act_func(z)
print(f"For Studyhours {test_value} Predicted value: {y_pred}")
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10,6))
plt.scatter(x,y,label="Data")
decision_boundary=-bias/w # as decision Boundary is w*x+bias=0 => x=-bias/w
plt.axvline(x=decision_boundary,c='red',label="decision boundary")
plt.xlabel("Study Hours")
plt.ylabel("Result Pass(1)/Fail(0)")
plt.legend()
plt.show()
```
