### Perceptron AND Gate

```python
import numpy as np
x=np.array([
    [0,0],
    [0,1],
    [1,0],
    [1,1]
])
y=np.array([0,0,0,1])
```

```python
len(x)
```

```python
w=np.zeros(2)
b=0.0
n=0.1 #learning Rate
epochs=10

print("Final Weight and Bias",w,b)
```

```python
def act_func(z):
    return 1 if z>=0 else 0
```

```python
for ep in range(epochs):
    print("")
    for i in range(len(x)):
        z=np.sum(x[i]*w)+b # weighted_sum
        # or
        # z=np.dot(x[i],w)+b
        y_pred=act_func(z)
        error=y[i]-y_pred
        w=w+n*error*x[i]
        b=b+n*error

        print(f"Input: {x[i]} Predicted value: {y_pred} weight: {w} Bias: {b}")
```

```python
print("Final Weight and Bias ",w,b)
```

```python
for i in range(len(x)):
    z=np.dot(w,x[i])+b
    y_pred=act_func(z)
    print(f"Input: {x[i]} Prediction: {y_pred}")
```

```python
# import matplotlib.pyplot as plt

import numpy as np
import matplotlib.pyplot as plt

# AND Gate Dataset
X = np.array([
    [0, 0],
    [0, 1],
    [1, 0],
    [1, 1]
])

y = np.array([0, 0, 0, 1])

# Example learned weights and bias for AND gate
w = np.array([1, 1])
bias = -1.5

plt.figure()

# Plot class 0 points
plt.scatter(X[y == 0][:, 0], X[y == 0][:, 1])

# Plot class 1 points
plt.scatter(X[y == 1][:, 0], X[y == 1][:, 1], marker='x')

# Decision boundary: w1*x1 + w2*x2 + b = 0
x_vals = np.linspace(-0.5, 1.5, 100)
y_vals = (-w[0]*x_vals - bias) / w[1]

plt.plot(x_vals, y_vals)

plt.xlim(-0.5, 1.5)
plt.ylim(-0.5, 1.5)
plt.xlabel("x1")
plt.ylabel("x2")
plt.title("Perceptron Decision Boundary (AND Gate)")
plt.grid()
plt.show()
```
