```python
import numpy as np
np.random.seed(0)
x=np.array([
    [120,85,150,95],# Glucose
    [25,30,35,22] # BMI
])
#Normalize i/p's
x=x/200
y=np.array([1,0,1,0]) # Risk of Diabetes
print(x.shape)
print(y.shape)
```

```python
def sigmoid(z):
    return 1/(1+np.exp(-z))
```

```python
# Hidden Layer (3 Neurons,2 Inputs)
w1=np.random.randn(3,2)*0.1
b1=np.zeros((3,1))
# O/P Layer (1 Neuron,3 Hidden Inputs)
w2=np.random.randn(1,3)*0.1
b2=np.zeros((1,1))
```

```python
w1,b1
```

```python
z1=np.dot(w1,x)+b1
a1=sigmoid(z1)
z2=np.dot(w2,a1)+b2
a2=sigmoid(z2)
print("Predicted Output:",a2)
```

```python
a1.T
```

```python
m=y.shape[0]
loss=-np.sum(y*np.log(a2+1e-8)+(1-y)*np.log(1-a2+1e-8))/m
loss
```

```python
dz2=a2-y
dw2=(1/m)*np.dot(dz2,a1.T)
db2=(1/m)*np.sum(dz2,axis=1,keepdims=True)
```

```python
da1=np.dot(w2.T,dz2)
dz1=da1*(1-a1)
dw1=(1/m)*np.dot(dz1,x.T)
db1=(1/m)*np.sum(dz1,axis=1,keepdims=True)
```

```python
learning_rate=1
w2-=learning_rate*dw2
b2-=learning_rate*db2

w1-=learning_rate*dw1
b1-=learning_rate*db1
```

```python
epoch=5000
for i in range(epoch):
    z1=np.dot(w1,x)+b1
    a1=sigmoid(z1)

    z2=np.dot(w2,a1)+b2
    a2=sigmoid(z2)

    loss=-np.sum(y*np.log(a2+1e-8)+(1-y)*np.log(1-a2+1e-8))/m

    dz2=a2-y
    dw2=(1/m)*np.dot(dz2,a1.T)
    db2=(1/m)*np.sum(dz2,axis=1,keepdims=True)

    da1=np.dot(w2.T,dz2)
    dz1=da1*(1-a1)

    dw1=(1/m)*np.dot(dz1,x.T)
    db1=(1/m)*np.sum(dz1,axis=1,keepdims=True)

    w2-=0.5*dw2
    b2-=0.5*db2
    w1-=0.5*dw1
    b1-=0.5*db1

    if i%100==0:
        print(f"epoch: {i} loss: {loss}")
```
