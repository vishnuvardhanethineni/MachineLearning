```python
import numpy as np
x=np.array([
    [0,0],
    [0,1],
    [1,0],
    [1,1]
])
y=np.array([0,1,1,0])
```

```python
w1=np.array([1,1])
b1=-0.5
```

```python
# h1=[np.dot(w1,x[i])+b1 for i in range(len(x))]
h1=np.dot(x,w1)+b1
```

```python
h1
```

```python
# print(h1)
w2=np.array([1,1])
b2=-1.5
h2=np.dot(x,w2)+b2 
```

```python
def act_func(z):
    return 1 if z>=0 else 0
    # return (z>=0).astype(int)
```

```python
y1=[act_func(z) for z in h1]
y2=[act_func(z) for z in h2]
```

```python
y1,y2
```

```python
w2=[1,-2]
b3=-0.5
y_weight=[w2[0]*y1[i]+w2[1]*y2[i]+b3 for i in range(len(y))]
```

```python
print(y_weight)
y_pred=[act_func(z) for z in y_weight]
loss=y-y_pred
print("Loss: ",loss)
```
