```python
import pandas as pd
df=pd.read_csv(r'C:\Users\VISHNU\Downloads\MachineLearning\MachineLearning\logistic_regression\heart.csv')
df.head()
```

```python
df.isnull().sum()
```

```python
x=df.drop(columns='target',axis=1)
y=df.target
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=2)
```

```python
from sklearn.preprocessing import LabelEncoder,StandardScaler
encoders={}
for c in x.select_dtypes(include='object').columns:
    le=LabelEncoder()
    x_train[c]=le.fit_transform(x_train[c])
    x_test[c]=le.transform(x_test)
    encoders[c]=le
    
```

```python
num=x_train.select_dtypes(include='number').columns
sc=StandardScaler()
x_train[num]=sc.fit_transform(x_train[num])
x_test[num]=sc.transform(x_test[num])
```

```python
from sklearn.svm import SVC
svm_linear=SVC(kernel='linear',C=1)
svm_linear.fit(x_train,y_train)
y_pred=svm_linear.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix
print("accuracy: ",accuracy_score(y_test,y_pred))
print("confusion matrix\n",confusion_matrix(y_test,y_pred))
```

```python
svm_poly=SVC(kernel='poly',degree=3,C=1)
svm_poly.fit(x_train,y_train)
y_pred=svm_poly.predict(x_test)
```

```python
print("accuracy_score: ",accuracy_score(y_test,y_pred))
print("confusion matrix\n",confusion_matrix(y_test,y_pred))
```
